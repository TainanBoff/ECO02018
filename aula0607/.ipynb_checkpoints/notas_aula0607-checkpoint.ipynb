## Importação dos módulos

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm # API padrao. Sintaxe do Python.
import statsmodels.formula.api as smf # As funcoes tem a sintaxe do R.
```

## Leitura das séries de dados no SGS/BCB.

```python
pib = pd.read_csv('http://api.bcb.gov.br/dados/serie/bcdata.sgs.22099/dados?formato=csv',
                  sep = ';', encoding = 'utf-8', decimal = ',')
consumo = pd.read_csv('http://api.bcb.gov.br/dados/serie/bcdata.sgs.22100/dados?formato=csv',
                  sep = ';', encoding = 'utf-8', decimal = ',')
consumoGoverno = pd.read_csv('http://api.bcb.gov.br/dados/serie/bcdata.sgs.22101/dados?formato=csv',
                  sep = ';', encoding = 'utf-8', decimal = ',')
investimento = pd.read_csv('http://api.bcb.gov.br/dados/serie/bcdata.sgs.22102/dados?formato=csv',
                  sep = ';', encoding = 'utf-8', decimal = ',')
exportacao = pd.read_csv('http://api.bcb.gov.br/dados/serie/bcdata.sgs.22103/dados?formato=csv',
                  sep = ';', encoding = 'utf-8', decimal = ',')
importacao = pd.read_csv('http://api.bcb.gov.br/dados/serie/bcdata.sgs.22104/dados?formato=csv',
                  sep = ';', encoding = 'utf-8', decimal = ',')
```

## Calculando as diferenças logarítmicas das séries

Retornos logaritmicos tem propriedades desejaveis para series temporais, como estacionariedade da media.

```python
dlogconsumo = np.log(consumo.valor).diff()
dlogconsumo.head()
```

```python
grafLogconsumo = plt.plot(dlogconsumo)
plt.savefig("logconsumo2.png")
plt.show()
```

```python
dlogpib = np.log(pib.valor).diff()
dlogpib.head()
```

```python
grafLogpib = plt.plot(dlogpib)
plt.savefig("logpib2.png")
plt.show()
```

## Efetuando a regressão por mínimos quadrados ordinários

```python
# Regressão
results = sm.OLS(dlogconsumo, dlogpib, missing = 'drop').fit()
print(results.summary())
```

#### Exercicio 1:

```python
adfconsumo = sm.tsa.stattools.adfuller(consumo.valor)
print("estatistica de teste:", adfconsumo[0])
print("p-valor:", adfconsumo[1])
print("numero de lags:", adfconsumo[2])
print("numero de observacoes", adfconsumo[3])
print("valores criticos para a estatistica de teste:", adfconsumo[4])
```

Conclusao: a estatistica de teste e maior do que os valores criticos tabelados. Logo, nao e possivel rejeitar a hipotese nula de raiz unitaria ao nivel de significancia de 10%. A serie nao e estacionaria.

```python
adfpib = sm.tsa.stattools.adfuller(pib.valor)
print("estatistica de teste:", adfpib[0])
print("p-valor:", adfpib[1])
print("numero de lags:", adfpib[2])
print("numero de observacoes", adfpib[3])
print("valores criticos para a estatistica de teste:", adfpib[4])
```
